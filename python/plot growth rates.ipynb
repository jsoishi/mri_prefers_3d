```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import numpy.polynomial.polynomial as poly
import h5py
```

```python
ssc = np.array([0.1, 0.2,0.3,0.35,0.4])
gamma = np.array([-2.64401117429893505188e-05,0.004275413721579234, 0.010448464864944,0.013961319783406,0.017668042014665])
```

```python
ssc_new = np.linspace(-0.1,ssc[-1],100)
```

```python
coefs_1 = poly.polyfit(ssc, gamma, 1)
ffit_1 = poly.polyval(ssc_new, coefs_1)
coefs_2 = poly.polyfit(ssc, gamma, 2)
ffit_2 = poly.polyval(ssc_new, coefs_2)
```

```python
plt.figure()
plt.plot(ssc, gamma,'o')
#plt.plot(ssc_new, ffit_1)
plt.plot(ssc_new, ffit_2)
plt.axhline(0,color='k',alpha=0.4)
plt.xlabel("$S/S_c$")
plt.ylabel("$\gamma$")
```
